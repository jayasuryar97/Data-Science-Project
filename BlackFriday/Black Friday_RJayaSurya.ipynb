# 1-Gender

```python
#Exploring the percentage of each gender on the Black Friday sales.
explode = (0.1,0)  
fig1, ax1 = plt.subplots(figsize=(12,7))
ax1.pie(bf['Gender'].value_counts(), explode=explode,labels=['Male','Female'], autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')  
plt.tight_layout()
plt.legend()
plt.show()
#plt.savefig("D:\Live_Project\BlackFridayGraphs\MaleandFemalePercentage.png",bbox_inches = 'tight')
```

```python
#Defining a plot function to group a certain column with respect to another column.
def plot(group,column,plot):
    ax=plt.figure(figsize=(12,6))
    bf.groupby(group)[column].sum().sort_values().plot(plot)

# Relation between the gender and the purchases.
plot('Gender','Purchase','bar')
#plt.savefig("D:\Live_Project\BlackFridayGraphs\GenderVsPurchase.png")
```

Men's purchasing power is greater than women's purchasing power, even in normal circumstances. This is likely to affect the owner of the money, but there has been a high turnout of men in the store.
About 75% of the customers have made sales of men of all ages, the men are generally heading toward products at 8000 - 12,000, we have probably made sales worth more than 4 billion in men and more than 1 billion in ladies.

# 3-City

```python
#Attendance of customers from each city.
explode = (0.1, 0, 0)
fig1, ax1 = plt.subplots(figsize=(12,7))
ax1.pie(bf['City_Category'].value_counts(),explode=explode, labels=bf['City_Category'].unique(), autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.legend()
plt.show()
#plt.savefig("D:\Live_Project\BlackFridayGraphs\CityCategory.png",bbox_inches = 'tight')
```

```python
#Purchases made by customers from each city.
explode = (0.1, 0, 0)
fig1, ax1 = plt.subplots(figsize=(12,7))
ax1.pie(bf.groupby('City_Category')['Purchase'].sum(),explode=explode, labels=bf['City_Category'].unique(), autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.legend()
plt.show()
#plt.savefig("D:\Live_Project\BlackFridayGraphs\CityCategoryPurchase.png",bbox_inches = 'tight')
```

```python
#Customer age groups from different cities.
fig1, ax1 = plt.subplots(figsize=(12,7))
sns.countplot(bf['City_Category'],hue=bf['Age'])
#plt.savefig("D:\Live_Project\BlackFridayGraphs\CityCategoryandAge.png")
```
